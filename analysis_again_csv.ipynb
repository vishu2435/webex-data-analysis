```python
import re
regexp = '\w*[" "a-z\D]*mi[tra]\w+|\w*[" "a-z\D]*res[ea]\w+|\w*[" "a-z\D]*sa[ta]\w+|\w*[" "a-z\D]*edi\w+|\w*[" "a-z\D]*dev\w+|\w*[" "a-z\D]*bl[og]\w+|\w*[" "a-z\D]*pla\w+|\w*[" "a-z\D]*vu\w+|\w*[" "a-z\D]*reg\w+|\w*[" "a-z\D]*phy\w+|\w*[" "a-z\D]*ent\w+'

updatedList = {}
for i in list_of_students['Roll-no']:
    updatedList[i] = []
    print(i)
    for j,k in enumerate(txtdata[0]):
#         print(j,k)
        
        k = str(k)
        if i in k:
            updatedList[i].append(re.findall(regexp,k,re.IGNORECASE))
#     print(updatedList)
```

```python
for i in updatedList.keys():
    updatedList[i] = [item.lower() for sublist in updatedList[i] for item in sublist]
#     print(len(updatedList[i]))
```

```python
for j,i in enumerate(updatedList.keys()):
    print(j+2,'.',i,':',updatedList[i])
    
```
