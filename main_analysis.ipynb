```python
import re
email_regex = "[0-9]+[\w+\d+]*@[\w\.\w]*"
students_intrested=[]
for i in df[0]:
#     print(i)
    i = str(i)
#     print(i)
#     print(re.findall(r"[0-9]+[\w+\d+]*@[\w\.\w]*",i,re.IGNORECASE))
    if i and len(re.findall(email_regex,i,re.IGNORECASE))>0:
        students_intrested.append(i)
```

```python
entrepreneurship = []
for i in students_intrested:
    if re.match('\w*[" "a-z\D]*ent\w+',i,re.IGNORECASE):
        print(i)
        entrepreneurship.append(i)
entrepreneurship.pop(0)
entrepreneurship.pop(0)
len(entrepreneurship)
entrepreneurship
```

```python
groups = {
    'mitr':mitr,
    'editorial':editorial,
    'developer':developer,
    'blog':blog,
    'placement wing':placement_wing,
    'vue':vue,
    'regional coordinator':regional_coordinator,
    'physical meetup':physical_meetup,
    'research':research,
    'saathi':saathi,
    'enterpreneurship':entrepreneurship
}
students_dict = {}
email_regex = "[0-9]+[\w+\d+]*@[\w\.\w]*"
for grp in groups.keys():
    for i in groups[grp]:
        x = re.findall(email_regex,i)
        print(x)
        if len(x)>0: 
            if x[0] in students_dict:
                students_dict[x[0]].add(grp)
            else:
                students_dict[x[0]] = {grp}
            
            
        
```

```python
len(students_dict.keys())
```

```python
for i in students_dict.keys():
    students_dict[i] = sorted(students_dict[i])
    

students_dict
```

```python
df = {}
for i in students_dict.keys():
    df[i]=[]
for i,j in students_dict.items():
    for grp in groups.keys():
#         print(i)
        if grp in j:
            df[i].append(grp)
        else:
            df[i].append("")
        
```

```python
dataf = pd.DataFrame(df).transpose()
dataf.columns = groups.keys()
```
